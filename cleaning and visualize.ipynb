```python
import pandas as pd

# Using raw string notation
file_path = r"C:\Users\donal\Desktop\Python\forbes\billionaires data.csv"

# Read the CSV file
df = pd.read_csv(file_path)

# df copy
df_copy = df.copy()

# fill in nah value
df_new = df.drop(labels=['rank', 'birthDate','title'], axis=1)
df_new.head()
mean_age = df_new['age'].mean()
df_new['age'] = df['age'].fillna(mean_age)
df_new['country'] = df['country'].fillna("No Country")

# Change datatype
df_new['age'] = df_new['age'].astype(int)
df_new['finalWorth'] = df_new['finalWorth'].astype(float)
df_new.rename(columns={"finalWorth" : "finalWorth (M)"})

df_new['industries'] = df_new['industries'].str.strip('[]').str.replace("'", '')
```
